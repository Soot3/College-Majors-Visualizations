# Graduate Outcomes Dataset - Data Visualisations

The data used shows job outcomes of individuals who graduated from American colleges between 2011 and 2012, collated from the [American Community Survey](https://www.census.gov/programs-surveys/acs/). The current dataset used has been cleaned by [FiveThirtyEight](https://github.com/fivethirtyeight/data/tree/master/college-majors).

Using this dataset, the following topics were explored; Popularity of certain majors, Earnings from these majors, Male dominance in certain sectors.

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
recent_grads = pd.read_csv("recent-grads.csv")
recent_grads.info()
```

Each row in the dataset represents a different major in college and contains information on gender diversity, employment rates, median salaries, and more. Here are some of the columns in the dataset:

 - Rank - Rank by median earnings (the dataset is ordered by this column).
 - Major_code - Major code.
 - Major - Major description.
 - Major_category - Category of major.
 - Total - Total number of people with major.
 - Sample_size - Sample size (unweighted) of full-time.
 - Men - Male graduates.
 - Women - Female graduates.
 - ShareWomen - Women as share of total.
 - Employed - Number employed.
 - Median - Median salary of full-time, year-round workers.
 - Low_wage_jobs - Number in low-wage service jobs.
 - Full_time - Number employed 35 hours or more.
 - Part_time - Number employed less than 35 hours.

Looking at the dataset, we have a total of 21 columns and 173 rows, with no null values except in the total, men, women and ShareWomen columns.

## Data Visualizations

### Scatter Plots

With scatter plots, we can see the relationship between two variables visually, this helps in understanding or confirming influences between two distinct variables. 

Here we will look at some variables in the dataset

```python
recent_grads.plot("Sample_size", "Median", title = "Scatter Plot of Sample size vs Median Earnings", kind = "scatter")
```

From our earlier explanation, the 'Sample_size' column is a representation of the number of full-time individuals in this major, while the 'Median' column refers to the median earnings or salary for the major in question.

As we can see in our plot there is no significant relationship between these two columns.

```python
recent_grads.plot("Sample_size", "Unemployment_rate", title = "Sample size vs Unemployment rate", kind = "scatter")
```

Here we look at unemployment rate and the sample size, there is also no significant relationship

```python
recent_grads.plot("Full_time", "Median", title = "No of Full-time graduates vs Median Earnings", kind = "scatter")
```

This plot shows no significant relationship between the number of full-time graduates and the median earnings for the major.

However, this might be because the values of the 'Full_time' column would be better analyzed as a proportion of the total number of graduates in the major.

```python
recent_grads["Share_Full_time"] = recent_grads["Full_time"] / recent_grads["Total"]

recent_grads.plot("Share_Full_time", "Median", title = "Share of Full-time graduates vs Median Earnings", kind = "scatter")
```

Here we can see a weak, positive relationship between the two variables confirming that majors with a higher number of full-time graduates had a relatively higher median salary.

```python
recent_grads.plot("ShareWomen", "Unemployment_rate", title = "Share of Women vs the Unemployment Rate", kind = "scatter")
```

In this plot, we have another relatively weak, positive relationship between the proportion of women in a major and the unemployment rate in the major. This is a relationship that should have other external influences but its relationship is confirmed.

The higher the share of women in a major the higher the unemployment rate in that major  

```python
recent_grads.plot("ShareWomen", "Median", title = "Share of Women vs Median Earnings", kind = "scatter")
```

There is a weak negative relationship between the two variables such that majors with a higher share of women have a relatively lower median salary

### Scatter matrix

Scatter matrixes combine both scatter plots and histograms into one grid of plots and allows us to explore potential relationships and distributions simultaneously

```python
from pandas.plotting import scatter_matrix
```

```python
scatter_matrix(recent_grads[["Sample_size","Median"]], figsize = (10,10))
```

```python
scatter_matrix(recent_grads[["Sample_size","Median", "Unemployment_rate"]], figsize = (10,10))
```

### Bar Plots

The dataset has already been ordered by Rank which is based on the median salary, so taking the first and the last ten majors let's look at the share of women in these majors.

```python
recent_grads[:10].plot.barh(x='Major', y='ShareWomen', legend=False, title='Share of Women in the Majors with Top Median Salary')
recent_grads[-10:].plot.barh(x='Major', y='ShareWomen', legend=False, title='Share of Women in the Majors with Bottom Median Salary')
```

As we can see the share of women is lesser in the majors with higher median salaries. All the majors in the bottom ten have a share of women that is above half of their population, with 2 majors having no men at all.

However, there is a suprise in the form of the Astronomy and Astrophysics major which has a top range median salary and a share of women in their population that is about half.

```python
recent_grads[:10].plot.barh(x='Major', y='Unemployment_rate', legend=False, title='Unemployment Rate in the Majors with Top Median Salary')
recent_grads[-10:].plot.barh(x='Major', y='Unemployment_rate', legend=False, title='Unemployment Rate in the Majors with Bottom Median Salary')
```

Nuclear Engineering and Clinical Psychology have the highest unemployment rates in the respective plots. Astronomy and Astrophysics has the second lowest unemployment rate that we can see, confirming its place as the best major for women in this dataset in terms of median salary and unemployment rate
